```python
def get_j_cat(pref:str)->tuple[int, str]:   
    if pref == 'ttbar':
        return (70, 'OTHER_TTBAR')
    elif pref == 'yyyyZ':
        return (95, 'f6_yyyyZ')
    elif pref == 'vvWW':
        return (37, 'vvWW')
    elif pref == 'eeWW':
        return (21, 'eeWW')
    elif pref == 'xxWW':
        return (96, 'f6_xxWW')
    elif pref == 'xxxxZ':
        return (97, 'f6_xxxxZ')
    elif pref == 'llWW':
        return (22, 'llWW')

for f in relevant_processes:
    if '6f' in f:
        pref = f.split('_')[-1]
        j, cat = get_j_cat(pref)
        
        i = 1
        for p in obj[f]:
            l = str(f'{i:02d}')
            print(f'f6_{pref}_{p} = {j}{l},')
            i += 1
```
