```python
import ROOT
import uuid
%jsroot on

file = ROOT.TFile.Open("zhh_AIDA.root", "READ")
tree = file.Get("KinFitQQ_NMC")

# Generates a random 6-character string so ROOT never re-uses the memory space
canvas_name = f"canvas_{uuid.uuid4().hex[:6]}"
canvas = ROOT.TCanvas(canvas_name, "ZHH Analysis", 1500, 800)
canvas.Divide(4, 2)

ROOT.gStyle.SetOptStat(0)  # Hide stats box

variables = [
    "FitErrorCode", "FitProbability", "FitChi2", "pullJetEnergy"
]

for i, var in enumerate(variables):
    canvas.cd(i + 1)
    if var in ["FitProbability", "FitChi2"]:
        ROOT.gPad.SetLogy(1)
    tree.Draw(var)

# For "Before and After" variables
# Z Mass
canvas.cd(5)
tree.SetLineColor(ROOT.kRed)
tree.Draw("ZMassBeforeFit >> ZBF", "")
ZBF = ROOT.gDirectory.Get("ZBF")
ZBF.SetTitle("Z Mass; Mass GeV; Events") 
tree.SetLineColor(ROOT.kBlue)
tree.Draw("ZMassAfterFit >> ZAF", "", "SAME") 
ZAF = ROOT.gDirectory.Get("ZAF")

ZBF.SetMaximum(max(ZBF.GetMaximum(), ZAF.GetMaximum()) * 1.1)

leg_z = ROOT.TLegend(0.70, 0.70, 0.90, 0.90)
leg_z.SetBorderSize(0)
leg_z.SetFillColor(ROOT.kWhite)
leg_z.SetFillStyle(1001)
leg_z.AddEntry(ZBF, "Before Fit", "l")
leg_z.AddEntry(ZAF, "After Fit", "l")
leg_z.Draw()

# HH Mass
canvas.cd(6)
tree.SetLineColor(ROOT.kRed)
tree.Draw("HHMassBeforeFit >> HBF", "")
HBF = ROOT.gDirectory.Get("HBF")
HBF.SetTitle("HH Mass; Mass GeV; Events") 
tree.SetLineColor(ROOT.kBlue)
tree.Draw("HHMassAfterFit >> HAF", "", "SAME") 
HAF = ROOT.gDirectory.Get("HAF")

HBF.SetMaximum(max(HBF.GetMaximum(), HAF.GetMaximum()) * 1.1)

leg_h = ROOT.TLegend(0.20, 0.70, 0.40, 0.90)
leg_h.SetBorderSize(0)
leg_h.SetFillColor(ROOT.kWhite)
leg_h.SetFillStyle(1001)
leg_h.AddEntry(HBF, "Before Fit", "l")
leg_h.AddEntry(HAF, "After Fit", "l")
leg_h.Draw()

# ZHH Mass
canvas.cd(7)
tree.SetLineColor(ROOT.kRed)
tree.Draw("ZHHMassBeforeFit >> ZHHBF", "")
ZHHBF = ROOT.gDirectory.Get("ZHHBF")
ZHHBF.SetTitle("ZHH Mass; Mass GeV; Events") 
tree.SetLineColor(ROOT.kBlue)
tree.Draw("ZHHMassAfterFit >> ZHHAF", "", "SAME") 
ZHHAF = ROOT.gDirectory.Get("ZHHAF")

ZHHBF.SetMaximum(max(ZHHBF.GetMaximum(), ZHHAF.GetMaximum()) * 1.1)

leg_zhh = ROOT.TLegend(0.20, 0.70, 0.40, 0.90)
leg_zhh.SetBorderSize(0)
leg_zhh.SetFillColor(ROOT.kWhite)
leg_zhh.SetFillStyle(1001)
leg_zhh.AddEntry(ZHHBF, "Before Fit", "l")
leg_zhh.AddEntry(ZHHAF, "After Fit", "l")
leg_zhh.Draw()
canvas.Draw()
```
