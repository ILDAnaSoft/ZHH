```python
import ROOT
import uuid
%jsroot on

file = ROOT.TFile.Open("zhh_27mar_soft.root", "READ")
tree_zhh = file.Get("KinFitQQ_ZHH")
tree_zzh = file.Get("KinFitQQ_ZZH") 

canvas_name = f"canvas_{uuid.uuid4().hex[:6]}"
canvas = ROOT.TCanvas(canvas_name, "ZHH Analysis", 1500, 800)
canvas.Divide(5, 2)

ROOT.gStyle.SetOptStat(0)

# Explicit configuration for every single pad
plot_configs = [
    {"type": "standard", "var": "FitErrorCode", "logy": False, "title": "Fit Error Code"},
    {"type": "standard", "var": "FitProbability", "logy": True, "title": "Fit Probability"},
    {"type": "standard", "var": "FitChi2", "logy": True, "title": "Fit Chi2"},
    {"type": "standard", "var": "pullJetEnergy", "logy": False, "title": "Jet Energy Pull"},
    {"type": "mass", "before": "ZMassBeforeFit", "after": "ZMassAfterFit", "title": "Z Mass"},
    {"type": "mass", "before": "HHMassBeforeFit", "after": "HHMassAfterFit", "title": "HH Mass"},
    {"type": "mass", "before": "ZHHMassBeforeFit", "after": "ZHHMassAfterFit", "title": "ZHH Mass"}
]

for i, config in enumerate(plot_configs):
    canvas.cd(i + 1)
    
    # Check if this pad needs a log scale
    if config.get("logy"):
        ROOT.gPad.SetLogy(1)
    else:
        ROOT.gPad.SetLogy(0)
        
    if config["type"] == "mass":
        # --- MASS VARIABLES (3 Lines) ---
        var_before = config["before"]
        var_after = config["after"]
        title = config["title"]
        
        # Raw Data (Black) # Before fit for both zhh and zzh should be the same, so we can just use zhh for the raw distribution
        tree_zhh.SetLineColor(ROOT.kBlack)
        tree_zhh.Draw(f"{var_before} >> raw_{i}", "")
        raw = ROOT.gDirectory.Get(f"raw_{i}")
        if raw: raw.SetTitle(f"{title}; Mass (GeV); Events")
        
        # zzh (Red)
        tree_zzh.SetLineColor(ROOT.kRed)
        tree_zzh.Draw(f"{var_after} >> zzh_{i}", "", "SAME")
        zzh = ROOT.gDirectory.Get(f"zzh_{i}")
        
        # zhh (Blue)
        tree_zhh.SetLineColor(ROOT.kBlue)
        tree_zhh.Draw(f"{var_after} >> zhh_{i}", "", "SAME")
        zhh = ROOT.gDirectory.Get(f"zhh_{i}")
        
        # Overshoot protection
        if raw and zzh and zhh:
            raw.SetMaximum(max(raw.GetMaximum(), zzh.GetMaximum(), zhh.GetMaximum()) * 1.2)
            
        leg_cmd = f"leg_{i} = ROOT.TLegend(0.25, 0.70, 0.50, 0.90)\n"
        leg_cmd += f"leg_{i}.SetBorderSize(0)\n"
        leg_cmd += f"leg_{i}.AddEntry(raw, 'No fit', 'l')\n"
        leg_cmd += f"leg_{i}.AddEntry(zzh, 'ZZH', 'l')\n"
        leg_cmd += f"leg_{i}.AddEntry(zhh, 'ZHH', 'l')\n"
        leg_cmd += f"leg_{i}.Draw()"
        exec(leg_cmd)
        
    elif config["type"] == "standard":
        # --- STANDARD VARIABLES (2 Lines) ---
        var = config["var"]
        title = config["title"]
        
        # zzh (Red)
        tree_zzh.SetLineColor(ROOT.kRed)
        tree_zzh.Draw(f"{var} >> zzh_{i}", "")
        zzh = ROOT.gDirectory.Get(f"zzh_{i}")
        if zzh: zzh.SetTitle(f"{title}; Value; Events")
        
        # zhh (Blue)
        tree_zhh.SetLineColor(ROOT.kBlue)
        tree_zhh.Draw(f"{var} >> zhh_{i}", "", "SAME")
        zhh = ROOT.gDirectory.Get(f"zhh_{i}")
        
        # Overshoot protection
        if zzh and zhh:
            zzh.SetMaximum(max(zzh.GetMaximum(), zhh.GetMaximum()) * 1.2)
            
        leg_cmd = f"leg_{i} = ROOT.TLegend(0.70, 0.75, 0.90, 0.90)\n"
        leg_cmd += f"leg_{i}.SetBorderSize(0)\n"
        leg_cmd += f"leg_{i}.AddEntry(zzh, 'ZZH', 'l')\n"
        leg_cmd += f"leg_{i}.AddEntry(zhh, 'ZHH', 'l')\n"
        leg_cmd += f"leg_{i}.Draw()"
        exec(leg_cmd)

canvas.Draw()
```

```python
import ROOT
import uuid
%jsroot on

file = ROOT.TFile.Open("zhh_30mar_btag.root", "READ")
tree_zhh = file.Get("KinFitQQ_ZHH")
tree_zzh = file.Get("KinFitQQ_ZZH") 
tree_zhhsoft = file.Get("KinFitQQ_ZHHsoft")
tree_zzhsoft = file.Get("KinFitQQ_ZZHsoft")

ROOT.gStyle.SetOptStat(0)
ROOT.gStyle.SetPadLeftMargin(0.15) 
ROOT.gStyle.SetPadBottomMargin(0.12) 
ROOT.gStyle.SetTitleYOffset(2)

# Explicit configuration for every single pad
plot_configs = [
    {"var": "FitErrorCode", "logy": False, "title": "Fit Error Code", "xtitle": "Error Code"},
    {"var": "FitProbability", "logy": True, "title": "Fit Probability", "bins": 100, "xmin": 0.0, "xmax": 1.0, "xtitle": "Probability"},
    {"var": "FitChi2", "logy": True, "title": "Fit Chi2", "xtitle": "#chi^{2}"},
    {"var": "pullJetEnergy", "logy": False, "title": "Jet Energy Pull", "bins": 100, "xmin": -200, "xmax": 200,"xtitle": "Pull"},
    {"var": "ZMassAfterFit", "logy": False, "title": "Z Mass", "xtitle": "Mass (GeV)"},
    {"var": "H1MassAfterFit", "logy": False, "title": "H1 Mass", "xtitle": "Mass (GeV)"},
    {"var": "H2MassAfterFit", "logy": False, "title": "H2 Mass", "xtitle": "Mass (GeV)"},
    {"var": "ZHHMassAfterFit", "logy": False, "title": "ZHH Mass", "xtitle": "Mass (GeV)"},
    {"var": "ISREnergyBeforeFit", "logy": False, "title": "ISR Energy (Prefit)", "xtitle": "Energy (GeV)", "bins": 100, "xmin": 0.0, "xmax": 150.0},
    {"var": "ISREnergyAfterFit", "logy": True, "title": "ISR Energy (Postfit)", "xtitle": "Energy (GeV)", "bins": 100, "xmin": 0.0, "xmax": 170.0},
]

canvas_zhh_name = f"canvas_{uuid.uuid4().hex[:6]}"
canvas_zhh = ROOT.TCanvas(canvas_zhh_name, "ZHH Analysis", 3200, 1200)
canvas_zhh.Divide(5, 2)

for i, config in enumerate(plot_configs):
    canvas_zhh.cd(i + 1)
    var = config["var"]
    title = config["title"]
    xtitle = config.get("xtitle", "")

    FEC_cut = "FitErrorCode == 0" if var != "FitErrorCode" else ""

    # Check if this pad needs a log scale
    if config.get("logy"):
        ROOT.gPad.SetLogy(1)
    else:
        ROOT.gPad.SetLogy(0)

    # Dynamically check if custom binning is defined in the config
    if "bins" in config and "xmin" in config and "xmax" in config:
        b = config["bins"]
        x1 = config["xmin"]
        x2 = config["xmax"]
        draw_soft = f"{var} >> zhhsoft_{i}({b}, {x1}, {x2})"
        draw_hard = f"{var} >> zhh_{i}({b}, {x1}, {x2})"
    else:
        # Fallback to letting ROOT automatically build the axis
        draw_soft = f"{var} >> zhhsoft_{i}"
        draw_hard = f"{var} >> zhh_{i}"

   # zhhsoft (Blue)
    tree_zhhsoft.Draw(draw_soft, FEC_cut)
    zhhsoft = ROOT.gDirectory.Get(f"zhhsoft_{i}")
    zhhsoft.SetTitle(f"{title}; {xtitle}; Events")
    zhhsoft.SetLineColor(ROOT.kBlue)

    # zhh (Red)
    tree_zhh.Draw(draw_hard, FEC_cut, "SAME")
    zhh = ROOT.gDirectory.Get(f"zhh_{i}")
    zhh.SetLineColor(ROOT.kRed)

# Overshoot protection
    if zhhsoft and zhh:
        zhhsoft.SetMaximum(max(zhh.GetMaximum(), zhhsoft.GetMaximum()) * 1.2)
            
    leg_cmd = f"leg_zhh_{i} = ROOT.TLegend(0.25, 0.70, 0.50, 0.90)\n"
    leg_cmd += f"leg_zhh_{i}.SetBorderSize(0)\n"
    leg_cmd += f"leg_zhh_{i}.AddEntry(zhh, 'ZHH', 'l')\n"
    leg_cmd += f"leg_zhh_{i}.AddEntry(zhhsoft, 'ZHHsoft', 'l')\n"
    leg_cmd += f"leg_zhh_{i}.Draw()"
    exec(leg_cmd)

canvas_zhh.Draw()

canvas_zzh_name = f"canvas_{uuid.uuid4().hex[:6]}"
canvas_zzh = ROOT.TCanvas(canvas_zzh_name, "ZZH Analysis", 3200, 1200)
canvas_zzh.Divide(5, 2)

for i, config in enumerate(plot_configs):
    canvas_zzh.cd(i + 1)
    var = config["var"]
    title = config["title"]

    if var == "H2MassAfterFit":
        title = "Z2 Mass"
    elif var == "ZHHMassAfterFit":
        title = "ZZH Mass"

    xtitle = config.get("xtitle", "Value")
    
    FEC_cut = "FitErrorCode == 0" if var != "FitErrorCode" else ""

    # Check if this pad needs a log scale
    if config.get("logy"):
        ROOT.gPad.SetLogy(1)
    else:
        ROOT.gPad.SetLogy(0)

    # Dynamically check if custom binning is defined in the config
    if "bins" in config and "xmin" in config and "xmax" in config:
        b = config["bins"]
        x1 = config["xmin"]
        x2 = config["xmax"]
        draw_soft = f"{var} >> zzhsoft_{i}({b}, {x1}, {x2})"
        draw_hard = f"{var} >> zzh_{i}({b}, {x1}, {x2})"
    else:
        # Fallback to letting ROOT automatically build the axis
        draw_soft = f"{var} >> zzhsoft_{i}"
        draw_hard = f"{var} >> zzh_{i}"

    # zzhsoft (Blue)
    tree_zzhsoft.Draw(draw_soft, FEC_cut)
    zzhsoft = ROOT.gDirectory.Get(f"zzhsoft_{i}")
    zzhsoft.SetTitle(f"{title}; {xtitle}; Events")
    zzhsoft.SetLineColor(ROOT.kBlue)

    # zzh (Red)
    tree_zzh.Draw(draw_hard, FEC_cut, "SAME")
    zzh = ROOT.gDirectory.Get(f"zzh_{i}")
    zzh.SetLineColor(ROOT.kRed)

# Overshoot protection
    if zzhsoft and zzh:
        zzhsoft.SetMaximum(max(zzh.GetMaximum(), zzhsoft.GetMaximum()) * 1.2)
            
    leg_cmd = f"leg_zzh_{i} = ROOT.TLegend(0.25, 0.70, 0.50, 0.90)\n"
    leg_cmd += f"leg_zzh_{i}.SetBorderSize(0)\n"
    leg_cmd += f"leg_zzh_{i}.AddEntry(zzh, 'ZZH', 'l')\n"
    leg_cmd += f"leg_zzh_{i}.AddEntry(zzhsoft, 'ZZHsoft', 'l')\n"
    leg_cmd += f"leg_zzh_{i}.Draw()"
    exec(leg_cmd)

canvas_zzh.Draw()
```
