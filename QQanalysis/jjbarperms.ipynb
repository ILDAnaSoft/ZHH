```python
import ROOT
import math
import uuid
%jsroot on

filename_jj = "/data/dust/user/tmanhong/zhh/AnalysisFinal45perms/550-qqhh-fast-perf/E550-Test.Pqqhh.Gwhizard-2_8_5.*.slcio.root"
tree_zhh_jj = ROOT.TChain("KinFitQQ_ZHH")
tree_zhh_jj.Add(filename_jj)

ROOT.gStyle.SetOptStat(0)
ROOT.gStyle.SetPadLeftMargin(0.15)
ROOT.gStyle.SetPadBottomMargin(0.12)
ROOT.gStyle.SetTitleYOffset(2)

run_id = uuid.uuid4().hex[:6]
canvas_jj_name = f"canvas_{run_id}"
canvas_jj = ROOT.TCanvas(canvas_jj_name, "q-tag comparison", 1000, 700)

T_values = [0.0, 0.01, 0.1, 0.4]
colors = [ROOT.kRed, ROOT.kGreen + 2, ROOT.kMagenta + 2, ROOT.kAzure + 1]
hists = []
leg = ROOT.TLegend(0.70, 0.70, 0.88, 0.88)
leg.SetBorderSize(0)

for i, t in enumerate(T_values):
    hist_name = f"h_category_{run_id}_T{t}"
    h = ROOT.TH1F(hist_name, "j#bar{j} Permutation Classes by Threshold; ; Number of Events", 4, 0.5, 4.5)
    h.SetLineColor(colors[i])
    h.SetLineWidth(3) # Made lines thicker so they are easy to see when overlapping
    h.SetMinimum(0)

    # Set the X-axis labels
    xaxis = h.GetXaxis()
    xaxis.SetBinLabel(1, "18 perms")
    xaxis.SetBinLabel(2, "30 perms")
    xaxis.SetBinLabel(3, "36 perms")
    xaxis.SetBinLabel(4, "45 perms")
    xaxis.SetLabelSize(0.055)
    hists.append(h)
    leg.AddEntry(h, f"T = {t}", "l")

for event in tree_zhh_jj:
    if event.FitErrorCode != 0:
        continue

    log_ratios = []
    for i in range(6):
        score_b = max(event.scoreB[i], 1e-10)
        score_bbar = max(event.scoreBbar[i], 1e-10)
        score_c = max(event.scoreC[i], 1e-10)
        score_cbar = max(event.scoreCbar[i], 1e-10)

        logB = math.log10(score_b / score_bbar)
        logC = math.log10(score_c / score_cbar)
        log_ratios.append((logB, logC))

    for t_idx, t_val in enumerate(T_values):
        quarks = 0
        antiquarks = 0
        ambiguous = 0

        for logB, logC in log_ratios:
            sign_tag = 0 # ambigious=0, q=1, qbar=-1

            if logB > t_val or logC > t_val:
                sign_tag = 1

            if logB < -t_val or logC < -t_val:
                if sign_tag == 1:
                    sign_tag = 0
                else: 
                    sign_tag = -1

            if sign_tag == 1: quarks += 1
            elif sign_tag == -1: antiquarks += 1
            else: ambiguous += 1

        if quarks == 3 or antiquarks == 3:
            hists[t_idx].Fill(1)
        elif quarks == 2 and antiquarks == 2:
            hists[t_idx].Fill(2)
        elif quarks == 2 or antiquarks == 2 or quarks == 4 or antiquarks == 4:
            hists[t_idx].Fill(3)
        else:
            hists[t_idx].Fill(4)

max_bin = max(h.GetMaximum() for h in hists)
hists[0].SetMaximum(max_bin * 1.3)

hists[0].Draw("HIST")
for i in range(1, len(hists)):
    hists[i].Draw("HIST SAME")

leg.Draw()
canvas_jj.Draw()

try:
    plots.extend(hists)
    plots.extend([leg, canvas_jj])
except NameError:
    plots = hists.copy()
    plots.extend([leg, canvas_jj])
```
