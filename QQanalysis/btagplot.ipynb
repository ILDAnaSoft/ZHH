```python
import ROOT
import uuid
%jsroot on

filename = "/data/dust/user/tmanhong/zhh/AnalysisFinal45perms/550-qqhh-fast-perf/E550-Test.Pqqhh.Gwhizard-2_8_5.*.slcio.root"

tree_zhh = ROOT.TChain("KinFitQQ_ZHH")
tree_fs = ROOT.TChain("FinalStates")
tree_zhh.Add(filename)
tree_fs.Add(filename)
tree_zhh.AddFriend(tree_fs, "FS")

canvas_btags = ROOT.TCanvas(f"c_eff_{uuid.uuid4().hex[:6]}", "Global Argmax C-Subtraction", 800, 600)
ROOT.gStyle.SetOptStat(0)
ROOT.gStyle.SetPadLeftMargin(0.15) 
ROOT.gStyle.SetPadBottomMargin(0.12) 
ROOT.gStyle.SetTitleYOffset(2)

h_4b = ROOT.TH1F("h_4b", "Sum of B-tags; #Sigma B-Tag score; Events", 100, 0, 6.5)
h_6b = ROOT.TH1F("h_6b", "Sum of B-tags; #Sigma B-Tag score; Events", 100, 0, 6.5)

h_6b.SetLineColor(ROOT.kRed)
h_6b.SetLineWidth(2)
h_4b.SetLineColor(ROOT.kBlue)
h_4b.SetLineWidth(2)

total_true_6b = 0
total_true_4b = 0
passed_6b = 0  
passed_4b = 0  

for event in tree_zhh:
    
    if event.n_b_from_higgs != 4:
        continue
        
    n_b_total = sum(1 for pdg in event.final_states if abs(pdg) == 5)
    if n_b_total not in [4, 6]:
        continue

    # Extract all 4 scores individually
    try:
        b_raw = [max(event.scoreB[i], 1e-10) for i in range(6)]
        bbar_raw = [max(event.scoreBbar[i], 1e-10) for i in range(6)]
        c_raw = [max(event.scoreC[i], 1e-10) for i in range(6)]
        cbar_raw = [max(event.scoreCbar[i], 1e-10) for i in range(6)]
    except IndexError:
        continue
    
    # Calculate the combined b-score (This mathematically equals your bmax sum!)
    b_combined = [b_raw[i] + bbar_raw[i] for i in range(6)]
    raw_sum = sum(b_combined)
    mod_sum = raw_sum
    
    # Check ALL 6 jets using Argmax logic
    for i in range(6):
        
        # Which of the 4 nodes fired the highest for THIS jet?
        highest_node = max(b_raw[i], bbar_raw[i], c_raw[i], cbar_raw[i])
        
        # If the tagger explicitly predicts c or cbar, discard its b-score from the sum
        if highest_node == c_raw[i] or highest_node == cbar_raw[i]:
            mod_sum -= b_combined[i]

    # Fill histograms
    if n_b_total == 6:
        h_6b.Fill(mod_sum)
        total_true_6b += 1
        if mod_sum >= 4.0:
            passed_6b += 1
            
    elif n_b_total == 4:
        h_4b.Fill(mod_sum)
        total_true_4b += 1
        if mod_sum >= 4.0:
            passed_4b += 1

h_4b.Draw("HIST")
h_6b.Draw("HIST SAME")

max_y = max(h_4b.GetMaximum(), h_6b.GetMaximum())
if max_y > 0:
    h_4b.SetMaximum(max_y * 1.2)

leg = ROOT.TLegend(0.15, 0.75, 0.45, 0.85)
leg.SetBorderSize(0)
leg.AddEntry(h_6b, "True 6 b-jets", "l")
leg.AddEntry(h_4b, "True 4 b-jets", "l")
leg.Draw()

canvas_btags.Draw()

if total_true_6b > 0:
    eff_6b = (passed_6b / total_true_6b) * 100
    print(f"True 6b: {eff_6b:.2f}%  ({passed_6b}/{total_true_6b})")
if total_true_4b > 0:
    rej_4b = ((total_true_4b - passed_4b) / total_true_4b) * 100
    print(f"True 4b: {rej_4b:.2f}%  ({total_true_4b - passed_4b}/{total_true_4b})")

try:
    plots.extend([h_4b, h_6b, leg, canvas_btags])
except NameError:
    plots = [h_4b, h_6b, leg, canvas_btags]
```

```python
import ROOT
import uproot
import awkward as ak
import numpy as np
import uuid
%jsroot on

filename = "/data/dust/user/tmanhong/zhh/AnalysisFinal/550-qqhh-fast-perf/E550-Test.Pqqhh.Gwhizard-2_8_5.*.slcio.root"

scores = [
    "scoreU", "scoreUbar", "scoreD", "scoreDbar", "scoreS", "scoreSbar", "scoreC", "scoreCbar", "scoreB", "scoreBbar", "scoreG"
]

read = scores + ["true_jet_pdgs"]
main_data = uproot.concatenate(f"{filename}:KinFitQQ_ZHH", expressions=read)

np_pdgs = main_data["true_jet_pdgs"].to_numpy()       
n_jets = ak.num(main_data["scoreB"]).to_numpy()       

valid_mask = np.arange(10) < n_jets[:, None]

true_pdgs = np_pdgs[valid_mask]
abs_pdgs = np.abs(true_pdgs)

true_flavors = {
    "True b": {"mask": abs_pdgs == 5,   "color": ROOT.kBlue},
    "True c": {"mask": abs_pdgs == 4,   "color": ROOT.kRed},
    "True s": {"mask": abs_pdgs == 3,   "color": ROOT.kGreen+2},
    "True u": {"mask": abs_pdgs == 2,   "color": ROOT.kMagenta},
    "True d": {"mask": abs_pdgs == 1,   "color": ROOT.kCyan+1},
    "True g": {"mask": true_pdgs == 21, "color": ROOT.kOrange+1}
}

canvases = []
legends = [] 
stacks = []
hists_memory_saver = [] 


for score in scores:
    c_name = f"c_{score}_{uuid.uuid4().hex[:4]}"
    c = ROOT.TCanvas(c_name, f"Distribution of {score}", 800, 600)
    c.SetLogy() 
    
    hs = ROOT.THStack(f"hs_{score}", f"{score} by True Jet Flavor")

    leg = ROOT.TLegend(0.65, 0.55, 0.88, 0.85)
    leg.SetBorderSize(0)
    leg.SetFillStyle(0)
    
    score_arr = ak.flatten(main_data[score], axis=None).to_numpy()
    
    for flav_name, flav_info in true_flavors.items():
        hist_name = f"h_{score}_{uuid.uuid4().hex[:6]}"
        h = ROOT.TH1F(hist_name, f"{score} by True Jet Flavor", 50, 0, 1)
        
        masked_scores = score_arr[flav_info["mask"]]
        
        counts, _ = np.histogram(masked_scores, bins=50, range=(0, 1))
        
        for i in range(50):
            h.SetBinContent(i + 1, float(counts[i]))
    
        h.SetLineColor(flav_info["color"])
        h.SetLineWidth(2)
        h.SetFillStyle(0) 
        h.SetStats(0)

        hs.Add(h)
        leg.AddEntry(h, flav_name, "l")
        
        hists_memory_saver.append(h)
            
    hs.Draw("NOSTACK HIST")
    
    hs.GetXaxis().SetTitle(f"{score}")
    hs.GetYaxis().SetTitle("Number of Jets")
    hs.GetYaxis().SetTitleOffset(1.1)
    hs.SetMinimum(0.5) 

    leg.Draw()
    
    canvases.append(c)
    legends.append(leg) 
    stacks.append(hs)

for c in canvases:
    c.Draw()
```

```python
import ROOT
import uproot
import awkward as ak
import numpy as np
%jsroot on

ROOT.TH1.AddDirectory(False) # Stops ROOT from hoarding histograms globally
if ROOT.gROOT.FindObject("c_conf"):
    ROOT.gROOT.FindObject("c_conf").Close()
if ROOT.gROOT.FindObject("h_conf"):
    ROOT.gROOT.FindObject("h_conf").Delete()

filename = "/data/dust/user/tmanhong/zhh/AnalysisFinal/550-qqhh-fast-perf/E550-Test.Pqqhh.Gwhizard-2_8_5.*.slcio.root"

scores = ["scoreB", "scoreBbar", "scoreC", "scoreCbar", "scoreS", "scoreSbar", "scoreU", "scoreUbar", "scoreD", "scoreDbar", "scoreG"]

expressions_to_read = scores + ["true_jet_pdgs"]
main_data = uproot.concatenate(f"{filename}:KinFitQQ_ZHH", expressions=expressions_to_read)

flat_scores = [ak.flatten(main_data[s], axis=None).to_numpy() for s in scores]
score_matrix = np.column_stack(flat_scores)
pred_idx = np.argmax(score_matrix, axis=1)

np_pdgs = main_data["true_jet_pdgs"].to_numpy()      
n_jets = ak.num(main_data["scoreB"]).to_numpy()       

valid_mask_2d = np.arange(10) < n_jets[:, None]

true_pdgs = np_pdgs[valid_mask_2d]

pdg_to_idx = {
    5: 10, -5: 9, 4: 8, -4: 7, 3: 6, -3: 5, 2: 4, -2: 3, 1: 2, -1: 1, 21: 0
}

true_idx = np.array([pdg_to_idx.get(p, -1) for p in true_pdgs])

valid_mask = true_idx != -1
pred_idx_clean = pred_idx[valid_mask]
true_idx_clean = true_idx[valid_mask]

conf_matrix, _, _ = np.histogram2d(pred_idx_clean, true_idx_clean, bins=11, range=[[0, 11],[0, 11]])

row_sums = conf_matrix.sum(axis=0)
row_sums = np.where(row_sums == 0, 1, row_sums)
conf_matrix_norm = conf_matrix / row_sums[np.newaxis, :]


h_conf = ROOT.TH2F("h_conf", "ParT Flavour Tagging Confusion Matrix;Predicted Flavour;True Flavour", 11, 0, 11, 11, 0, 11)

for x in range(11):
    for y in range(11):
        h_conf.SetBinContent(x + 1, y + 1, conf_matrix_norm[x, y])

## ILD STYLING
if not ROOT.gROOT.GetStyle("ildStyle"):
    ildStyle = ROOT.TStyle("ildStyle", "ILD Style")
    ildStyle.SetFillColor(0)
    ildStyle.SetFrameFillColor(0)
    ildStyle.SetCanvasColor(0)
    ildStyle.SetPadColor(0)
    ildStyle.SetTitleFillColor(0)
    ildStyle.SetStatColor(0)
    ildStyle.SetFrameBorderMode(0)
    ildStyle.SetCanvasBorderMode(0)
    ildStyle.SetPadBorderMode(0)
    ildStyle.SetLegendBorderSize(0)

    ildStyle.SetPalette(ROOT.kBird) 
    ildStyle.SetHistLineColor(ROOT.kBlack)
    ildStyle.SetLabelColor(ROOT.kBlack, "xyz")
    ildStyle.SetTitleColor(ROOT.kBlack)

    ildStyle.SetPadBottomMargin(0.18)
    ildStyle.SetPadTopMargin(0.08)
    ildStyle.SetPadRightMargin(0.12)
    ildStyle.SetPadLeftMargin(0.17)

    ildStyle.SetLabelFont(42, "xyz")
    ildStyle.SetLabelSize(0.06, "xyz")
    ildStyle.SetLabelOffset(0.015, "xyz")
    ildStyle.SetTitleFont(42, "xyz")
    ildStyle.SetTitleSize(0.07, "xyz")
    ildStyle.SetTitleOffset(1.1, "yz")
    ildStyle.SetTitleOffset(1.0, "x")
    ildStyle.SetOptStat(0)
    ildStyle.SetOptTitle(1) 
    ildStyle.SetTitleBorderSize(0)  
    ildStyle.SetTitleAlign(23)     
    ildStyle.SetTitleX(0.5)         

ROOT.gStyle.SetPaintTextFormat(".3f")
ROOT.gROOT.SetStyle("ildStyle")
ROOT.gROOT.ForceStyle()

x_labels = ["b", "#bar{b}", "c", "#bar{c}", "s", "#bar{s}", "u", "#bar{u}", "d", "#bar{d}", "g"]
y_labels = list(reversed(x_labels))

for i in range(11):
    h_conf.GetXaxis().SetBinLabel(i + 1, x_labels[i])
    h_conf.GetYaxis().SetBinLabel(i + 1, y_labels[i])

h_conf.GetXaxis().SetLabelSize(0.05)
h_conf.GetYaxis().SetLabelSize(0.05)
h_conf.GetZaxis().SetLabelSize(0.03)
h_conf.GetXaxis().SetTitleSize(0.03)
h_conf.GetYaxis().SetTitleSize(0.03)
h_conf.GetZaxis().SetTitleSize(0.03)
h_conf.GetXaxis().SetTitleOffset(1.6)  
h_conf.GetYaxis().SetTitleOffset(1.3)

c_conf = ROOT.TCanvas("c_conf", "Confusion Matrix", 900, 700)

h_conf.Draw("COLZ")
text_objects = []

# Loop over the 11x11 bins (ROOT bins are 1-indexed)
for x in range(1, 12):
    for y in range(1, 12):
        val = h_conf.GetBinContent(x, y)
        
        # The kBird palette is dark blue at the bottom and bright yellow at the top.
        # A threshold of ~0.4 or 0.5 works perfectly to swap the text color.
        text_color = ROOT.kBlack if val > 0.2 else ROOT.kWhite
        
        # Get the geometric center of the current bin
        x_center = h_conf.GetXaxis().GetBinCenter(x)
        y_center = h_conf.GetYaxis().GetBinCenter(y)
        
        # Format the text (using 4 decimal places as seen in your previous plot)
        val_str = format(val, ".4f")
        
        # Draw the text manually
        latex = ROOT.TLatex(x_center, y_center, val_str)
        latex.SetTextAlign(22)          # Center align both horizontally and vertically
        latex.SetTextFont(42)           # Match your ILD style font
        latex.SetTextSize(0.022)        # Tweak this size to fit the boxes properly
        latex.SetTextColor(text_color)
        
        latex.Draw()
        text_objects.append(latex)      # Store it so it actually renders!


c_conf.Draw()
```

```python
import ROOT
import uuid
%jsroot on

file = "/data/dust/user/tmanhong/zhh/AnalysisFinalPrejjbarperms/550-qqhh-fast-perf/E550-Test.Pqqhh.Gwhizard-2_8_5.*.slcio.root"

tree_zhh = ROOT.TChain("KinFitQQ_ZHH")
tree_zzh = ROOT.TChain("KinFitQQ_ZZH") 
tree_zhhsoft = ROOT.TChain("KinFitQQ_ZHHsoft")
tree_zzhsoft = ROOT.TChain("KinFitQQ_ZZHsoft")

tree_zhh.Add(file)
tree_zzh.Add(file)
tree_zhhsoft.Add(file)
tree_zzhsoft.Add(file)

ROOT.gStyle.SetOptStat(0)
ROOT.gStyle.SetPadLeftMargin(0.15) 
ROOT.gStyle.SetPadBottomMargin(0.12) 
ROOT.gStyle.SetTitleYOffset(2)

# Explicit configuration for every single pad
plot_configs = [
    {"var": "FitErrorCode", "logy": False, "title": "Fit Error Code", "xtitle": "Error Code"},
    {"var": "FitProbability", "logy": True, "title": "Fit Probability", "bins": 100, "xmin": 0.0, "xmax": 1.0, "xtitle": "Probability"},
    {"var": "FitChi2", "logy": True, "title": "Fit Chi2", "xtitle": "#chi^{2}", "bins": 100, "xmin": 0.0, "xmax": 1e10},
    {"var": "pullJetEnergy", "logy": False, "title": "Jet Energy Pull", "bins": 100, "xmin": -50, "xmax": 50,"xtitle": "Pull"},
    {"var": "ZMassAfterFit", "logy": False, "title": "Z Mass", "xtitle": "Mass (GeV)", "bins": 100, "xmin": 0.0 , "xmax": 100},
    {"var": "H1MassAfterFit", "logy": False, "title": "H1 Mass", "xtitle": "Mass (GeV)", "bins": 100, "xmin": 0.0, "xmax": 150.0},
    {"var": "H2MassAfterFit", "logy": False, "title": "H2 Mass", "xtitle": "Mass (GeV)", "bins": 100, "xmin": 0.0, "xmax": 150.0},
    {"var": "ZHHMassAfterFit", "logy": False, "title": "ZHH Mass", "xtitle": "Mass (GeV)", "bins": 100, "xmin": 0.0, "xmax": 600.0},
    {"var": "ISREnergyBeforeFit", "logy": False, "title": "ISR Energy (Prefit)", "xtitle": "Energy (GeV)", "bins": 100, "xmin": 0.0, "xmax": 150.0},
    {"var": "ISREnergyAfterFit", "logy": True, "title": "ISR Energy (Postfit)", "xtitle": "Energy (GeV)", "bins": 100, "xmin": 0.0, "xmax": 170.0},
]

# Create a master list to store legends from BOTH canvases so they don't disappear
legends = []

# ==========================================
# Canvas 1: ZHH Analysis
# ==========================================
canvas_zhh_name = f"canvas_{uuid.uuid4().hex[:6]}"
canvas_zhh = ROOT.TCanvas(canvas_zhh_name, "ZHH Analysis", 3200, 1200)
canvas_zhh.Divide(5, 2)

for i, config in enumerate(plot_configs):
    canvas_zhh.cd(i + 1)
    var = config["var"]
    title = config["title"]
    xtitle = config.get("xtitle", "")

    FEC_cut = "FitErrorCode == 0" if var != "FitErrorCode" else ""

    ROOT.gPad.SetLogy(1 if config.get("logy") else 0)

    if "bins" in config and "xmin" in config and "xmax" in config:
        b, x1, x2 = config["bins"], config["xmin"], config["xmax"]
        draw_soft = f"{var} >> zhhsoft_{i}({b}, {x1}, {x2})"
        draw_hard = f"{var} >> zhh_{i}({b}, {x1}, {x2})"
    else:
        draw_soft = f"{var} >> zhhsoft_{i}"
        draw_hard = f"{var} >> zhh_{i}"

    # zhhsoft (Blue)
    tree_zhhsoft.Draw(draw_soft, FEC_cut)
    zhhsoft = ROOT.gDirectory.Get(f"zhhsoft_{i}")
    zhhsoft.SetTitle(f"{title}; {xtitle}; Events")
    zhhsoft.SetLineColor(ROOT.kBlue)

    # zhh (Red)
    tree_zhh.Draw(draw_hard, FEC_cut, "SAME")
    zhh = ROOT.gDirectory.Get(f"zhh_{i}")
    zhh.SetLineColor(ROOT.kRed)

    # Overshoot protection
    if zhhsoft and zhh:
        zhhsoft.SetMaximum(max(zhh.GetMaximum(), zhhsoft.GetMaximum()) * 1.2)
            
    # FIXED: Clean Pythonic Legend creation
    leg = ROOT.TLegend(0.25, 0.70, 0.50, 0.90)
    leg.SetBorderSize(0)
    leg.AddEntry(zhh, 'ZHH', 'l')
    leg.AddEntry(zhhsoft, 'ZHHsoft', 'l')
    leg.Draw()
    legends.append(leg)

canvas_zhh.Draw()

# ==========================================
# Canvas 2: ZZH Analysis
# ==========================================
canvas_zzh_name = f"canvas_{uuid.uuid4().hex[:6]}"
canvas_zzh = ROOT.TCanvas(canvas_zzh_name, "ZZH Analysis", 3200, 1200)
canvas_zzh.Divide(5, 2)

for i, config in enumerate(plot_configs):
    canvas_zzh.cd(i + 1)
    var = config["var"]
    title = config["title"]

    # Smart title replacement for ZZH
    if var == "H2MassAfterFit":
        title = "Z2 Mass"
    elif var == "ZHHMassAfterFit":
        title = "ZZH Mass"

    xtitle = config.get("xtitle", "Value")
    
    FEC_cut = "FitErrorCode == 0" if var != "FitErrorCode" else ""

    ROOT.gPad.SetLogy(1 if config.get("logy") else 0)

    if "bins" in config and "xmin" in config and "xmax" in config:
        b, x1, x2 = config["bins"], config["xmin"], config["xmax"]
        draw_soft = f"{var} >> zzhsoft_{i}({b}, {x1}, {x2})"
        draw_hard = f"{var} >> zzh_{i}({b}, {x1}, {x2})"
    else:
        draw_soft = f"{var} >> zzhsoft_{i}"
        draw_hard = f"{var} >> zzh_{i}"

    # zzhsoft (Blue)
    tree_zzhsoft.Draw(draw_soft, FEC_cut)
    zzhsoft = ROOT.gDirectory.Get(f"zzhsoft_{i}")
    zzhsoft.SetTitle(f"{title}; {xtitle}; Events")
    zzhsoft.SetLineColor(ROOT.kBlue)

    # zzh (Red)
    tree_zzh.Draw(draw_hard, FEC_cut, "SAME")
    zzh = ROOT.gDirectory.Get(f"zzh_{i}")
    zzh.SetLineColor(ROOT.kRed)

    # Overshoot protection
    if zzhsoft and zzh:
        zzhsoft.SetMaximum(max(zzh.GetMaximum(), zzhsoft.GetMaximum()) * 1.2)
            
    # FIXED: Clean Pythonic Legend creation
    leg = ROOT.TLegend(0.25, 0.70, 0.50, 0.90)
    leg.SetBorderSize(0)
    leg.AddEntry(zzh, 'ZZH', 'l')
    leg.AddEntry(zzhsoft, 'ZZHsoft', 'l')
    leg.Draw()
    legends.append(leg)

canvas_zzh.Draw()
```

```python
%jsroot on
import ROOT

# Load the file and tree
file = ROOT.TFile("zhh_AIDA.root", "READ")
tree = file.Get("KinFitQQ_ZHH")

# Define the 11 score branches you want to plot
scores = [
    "scoreU", "scoreUbar", 
    "scoreD", "scoreDbar", 
    "scoreS", "scoreSbar", 
    "scoreC", "scoreCbar", 
    "scoreB", "scoreBbar", 
    "scoreG"
]

# Define the 6 channels using the Sum$ ROOT trick
channels = {
    "uubbbb": {"cut": "Sum$(abs(true_jet_pdgs)==2)==2", "color": ROOT.kMagenta},
    "ddbbbb": {"cut": "Sum$(abs(true_jet_pdgs)==1)==2", "color": ROOT.kCyan+1},
    "ssbbbb": {"cut": "Sum$(abs(true_jet_pdgs)==3)==2", "color": ROOT.kGreen+2},
    "ccbbbb": {"cut": "Sum$(abs(true_jet_pdgs)==4)==2", "color": ROOT.kRed},
    "bbbbbb": {"cut": "Sum$(abs(true_jet_pdgs)==5)==6", "color": ROOT.kBlue},
    "ggbbbb": {"cut": "Sum$(true_jet_pdgs==21)==2",      "color": ROOT.kOrange+1}
}

# Lists to prevent PyROOT garbage collection!
canvases = []
legends = [] 
hists_dict = {}

for score in scores:
    c = ROOT.TCanvas(f"c_{score}", f"Distribution of {score}", 800, 600)
    c.SetLogy() 
    
    hists_dict[score] = []
    max_y = 0
    
    for chan_name, chan_info in channels.items():
        hist_name = f"h_{score}_{chan_name}"
        
        h = ROOT.TH1F(hist_name, f"Distribution of {score} by Decay Channel", 50, 0, 1)
        tree.Draw(f"{score} >> {hist_name}", chan_info["cut"], "GOFF")
        
        h.SetLineColor(chan_info["color"])
        h.SetLineWidth(2)
        h.SetStats(0)
        h.GetXaxis().SetTitle(f"{score}")
        h.GetYaxis().SetTitle("Number of Jets")
        h.SetMinimum(0.5) 
        
        if h.GetMaximum() > max_y:
            max_y = h.GetMaximum()
            
        hists_dict[score].append((h, chan_name))
    
    # Create the legend and adjust its size/position
    leg = ROOT.TLegend(0.65, 0.55, 0.88, 0.85)
    leg.SetBorderSize(0)
    leg.SetFillStyle(0) # Transparent background
    
    first = True
    for h, chan_name in hists_dict[score]:
        if first:
            h.SetMaximum(max_y * 20) # Extra headroom for the bigger legend
            h.Draw("HIST")
            first = False
        else:
            h.Draw("HIST SAME")
            
        leg.AddEntry(h, f"{chan_name} events", "l")
        
    leg.Draw()
    
    # Save the canvas AND the legend to the persistent lists
    canvases.append(c)
    legends.append(leg) 

# Draw all canvases to the Jupyter output
for c in canvases:
    c.Draw()
```
